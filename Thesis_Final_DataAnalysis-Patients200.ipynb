```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Replace 'path_to_file' with the actual path to your CSV file
df = pd.read_csv('D:/UpGrad_Notes/ljmu immersion/Research/interim report/Coding/nifti/Patient200Master.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
unique_study_uids = df['Study UID'].nunique()
print("Number of unique Study UID:", unique_study_uids)
```

```python
df['age_modified'] = df['age'].str.extract('(\d+)').astype(int)
```

```python

df_agecount=df.drop_duplicates(subset=['age_modified','Study UID'])
```

```python
df_agecount.head()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df_agecount['age_modified'], kde=True, bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.xticks(rotation=45)  # Rotate the x-axis labels by 45 degrees
plt.xticks(fontsize=8)
plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
df_diag=df.drop_duplicates(subset=['diagnosis','Study UID'])
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df_diag['diagnosis'], kde=True, bins=20)
plt.xlabel('diagnosis')
plt.ylabel('Count')
plt.title('diagnosis Distribution')
plt.xticks(rotation=45)  # Rotate the x-axis labels by 45 degrees
plt.xticks(fontsize=8)
plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
df_sex=df.drop_duplicates(subset=['sex','age_modified','Study UID'])
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df_sex['sex'], kde=True, bins=20)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Sex Distribution')
plt.xticks(rotation=45)  # Rotate the x-axis labels by 45 degrees
plt.xticks(fontsize=8)
plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df_sex['diagnosis'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Diagnosis Distribution')
plt.ylabel('')
plt.show()
```

```python
df_sexunique=df.drop_duplicates(subset=['sex','age_modified','diagnosis','Study UID'])
```

```python
df_sexunique
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age_modified', y='diagnosis', data=df_sexunique)
plt.xlabel('Age')
plt.ylabel('Diagnosis')
plt.title('Diagnosis vs. Age Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='diagnosis', y='age_modified', data=df_sexunique)
plt.xlabel('Diagnosis')
plt.ylabel('Age')
plt.title('Age Distribution by Diagnosis')
plt.show()
```

```python
average_age_by_diagnosis = df_sexunique.groupby('diagnosis')['age_modified'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(x='diagnosis', y='age_modified', data=average_age_by_diagnosis)
plt.xlabel('Diagnosis')
plt.ylabel('Average Age')
plt.title('Average Age by Diagnosis')
plt.show()
```
