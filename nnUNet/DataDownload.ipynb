```python
import boto3
import os
import zipfile

# Create an S3 client
s3 = boto3.client('s3')

# Specify the bucket name and local directory to save files
bucket_name = 'pet-ct-lesions-s3'
local_directory = 'data'

# List objects in the bucket
objects = s3.list_objects_v2(Bucket=bucket_name)

# Download and unzip each file
for obj in objects.get('Contents', []):
    file_name = obj['Key']
    if file_name.endswith('.zip'):
        # Determine the name of the extraction folder
        extract_folder = os.path.join(local_directory, os.path.splitext(file_name)[0])
        
        # If the folder already exists, skip the download and extraction
        if os.path.exists(extract_folder):
            print(f"Folder '{extract_folder}' already exists. Skipping download and extraction.")
            continue
        
        local_file_path = os.path.join(local_directory, file_name)
        
        # Download the zip file
        s3.download_file(bucket_name, file_name, local_file_path)
        
        # Create the extraction folder
        os.makedirs(extract_folder, exist_ok=True)
        
        # Unzip the file into the new directory
        with zipfile.ZipFile(local_file_path, 'r') as zip_ref:
            zip_ref.extractall(extract_folder)
        
        # Optionally, remove the downloaded zip file after extraction
        os.remove(local_file_path)
```

```python
import os

def count_top_level_folders(directory):
    # List all items in the directory
    items = os.listdir(directory)
    
    # Filter out only the top-level folders (exclude files and nested folders)
    folders = [item for item in items if os.path.isdir(os.path.join(directory, item))]
    
    return len(folders)

# Specify the directory name (in this case, 'ZIP')
directory_name = 'data/ZIP'
folder_count = count_top_level_folders(directory_name)
print(f"'{directory_name}' contains {folder_count} top-level folders.")
```
